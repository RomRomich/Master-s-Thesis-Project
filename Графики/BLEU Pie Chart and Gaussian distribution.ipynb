```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv('data.csv')

def plot_and_print_bleu_by_category(direction):
    # Фильтрация по направлению перевода
    filtered_df = df[df['Направление перевода'] == direction]
    # Группировка по категориям и вычисление среднего значения BLEU, затем сортировка
    category_means = filtered_df.groupby('Категория')['BLEU'].mean().sort_values()
    # Печать значений от меньшего к большему
    print(f"\nСредние значения BLEU по категориям ({direction}) от меньшего к большему:")
    for category, value in category_means.items():
        print(f"{category}: {value:.4f}")
    # Построение круговой диаграммы
    plt.figure(figsize=(10, 8))
    plt.pie(category_means, labels=category_means.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Среднее значение BLEU по категориям ({direction})')
    plt.show()

# Построение диаграмм и печать значений для EN-RU и RU-EN
plot_and_print_bleu_by_category('EN-RU')
plot_and_print_bleu_by_category('RU-EN')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
df = pd.read_csv('data.csv')

# Функция для группировки и построения нормального распределения
def plot_gaussian_distribution_by_category(direction):
    # Фильтрация по направлению перевода
    filtered_df = df[df['Направление перевода'] == direction]
    # Группировка по категориям и вычисление среднего значения BLEU
    category_means = filtered_df.groupby('Категория')['BLEU'].mean()
    # Печать значений от меньшего к большему
    print(f"\nСредние значения BLEU по категориям ({direction}) от меньшего к большему:")
    sorted_category_means = category_means.sort_values()
    for category, value in sorted_category_means.items():
        print(f"{category}: {value:.4f}")
    # Построение графика с гауссовым распределением (кривой плотности KDE)
    plt.figure(figsize=(10, 6))
    sns.kdeplot(sorted_category_means, shade=True, bw_adjust=1.5)
    plt.title(f'Гауссово распределение BLEU по категориям ({direction})')
    plt.xlabel('BLEU Score')
    plt.ylabel('Density')
    plt.show()

# Построение диаграмм и печать значений для EN-RU и RU-EN
plot_gaussian_distribution_by_category('EN-RU')
plot_gaussian_distribution_by_category('RU-EN')
```
