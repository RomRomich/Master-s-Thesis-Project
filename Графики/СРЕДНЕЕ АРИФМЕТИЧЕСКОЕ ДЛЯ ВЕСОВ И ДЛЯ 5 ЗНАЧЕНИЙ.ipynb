КАТЕГОРИИ EN-RU И RU-EN ДЛЯ ВЕСОВ

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv('data.csv')

# Функция для расчета среднего по категориям и построения круговой диаграммы
def compute_and_plot_category_means(direction):
    # Фильтрация по направлению перевода
    filtered_df = df[df['Направление перевода'] == direction]

    # Группировка по категориям и вычисление среднего значения
    category_means = filtered_df.groupby('Категория')[['BLEU', 'ROUGE-1', 'METEOR']].mean()
    avg_means = category_means.mean(axis=1).sort_values()

    print(f"\nСредние значения по категориям для {direction}:")
    for index, value in avg_means.items():
        print(f"{index}: Среднее={value}")

    # Построение круговой диаграммы
    plt.figure(figsize=(10, 8))
    plt.pie(avg_means, labels=avg_means.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Средние значения по категориям для {direction}')
    plt.show()

# Средние значения и график для EN-RU
compute_and_plot_category_means('EN-RU')

# Средние значения и график для RU-EN
compute_and_plot_category_means('RU-EN')
```
