```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Загрузка данных из файла data.csv
df = pd.read_csv('data.csv')

# Перечисление метрик для анализа
metrics = ['BLEU', 'ROUGE-1', 'METEOR']

# Получение уникальных категорий
categories = df['Категория'].unique()

# Функция для построения графиков по категориям
def plot_metrics_by_category(df, category):
    subset = df[df['Категория'] == category]

    directions = subset['Направление перевода'].unique()
    mean_values = {}

    for direction in directions:
        mean_values[direction] = subset[subset['Направление перевода'] == direction][metrics].mean()
    
    fig, ax = plt.subplots(figsize=(12, 8))

    # Создание отдельных столбиков для каждого направления перевода
    bar_width = 0.35
    index = np.arange(len(metrics))
    colors = ['#1f77b4', '#2ca02c']  # Синий и зелёный

    for i, direction in enumerate(directions):
        values = mean_values[direction]
        bars = ax.bar(index + i * bar_width, values, width=bar_width, label=direction, color=colors[i], edgecolor='black')

        # Добавление текстовых меток над столбиками с обводкой
        for bar, value in zip(bars, values):
            ax.text(bar.get_x() + bar.get_width() / 2, value, f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='white',
                    bbox=dict(facecolor='black', alpha=0.7, edgecolor='none', boxstyle='round,pad=0.3'))

    ax.set_xlabel('Метрика', fontsize=14)
    ax.set_ylabel('Среднее значение', fontsize=14)
    ax.set_title(f'{category} - Среднее арифметическое значение метрик', fontsize=16)
    ax.set_xticks(index + bar_width / 2)
    ax.set_xticklabels(metrics, fontsize=12)
    ax.legend(title='Направление', fontsize=12, title_fontsize='13')

    # Убираем верхнюю и правую границы графика
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.grid(axis='y', linestyle='--', linewidth=0.7)
    plt.tight_layout()
    plt.show()

    # Вывод значений под графиком
    print(f"Значения для категории '{category}':")
    for direction, values in mean_values.items():
        sorted_values = values.sort_values()
        print(f"{direction}:")
        print(sorted_values)
    print("\n")

# Построение графика для каждой категории
for category in categories:
    plot_metrics_by_category(df, category)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Загрузка данных из файла data.csv
df = pd.read_csv('data.csv')

# Перечисление метрик для анализа
metrics = ['BLEU', 'ROUGE-1', 'METEOR']

# Получение уникальных категорий
categories = df['Категория'].unique()

# Функция для построения графиков по категориям
def plot_metrics_by_category(df, category):
    subset = df[df['Категория'] == category]

    directions = subset['Направление перевода'].unique()
    mean_values = {}

    for direction in directions:
        mean_values[direction] = subset[subset['Направление перевода'] == direction][metrics].mean()
    
    fig, ax = plt.subplots(figsize=(10, 6))

    # Создание отдельных столбиков для каждого направления перевода
    bar_width = 0.35
    index = np.arange(len(metrics))
    colors = ['blue', 'orange']
    
    for i, direction in enumerate(directions):
        values = mean_values[direction]
        ax.bar(index + i * bar_width, values, width=bar_width, label=direction, color=colors[i])

        # Добавление текстовых меток над столбиками
        for j, value in enumerate(values):
            ax.text(index[j] + i * bar_width, value, f'{value:.2f}', ha='center', va='bottom')

    ax.set_xlabel('Метрика')
    ax.set_ylabel('Среднее значение')
    ax.set_title(f'{category} - Среднее арифметическое значение метрик')
    ax.set_xticks(index + bar_width / 2)
    ax.set_xticklabels(metrics)
    ax.legend(title='Направление')

    plt.tight_layout()
    plt.show()

    # Вывод значений под графиком
    print(f"Значения для категории '{category}':")
    for direction, values in mean_values.items():
        sorted_values = values.sort_values()
        print(f"{direction}:")
        print(sorted_values)
    print("\n")

# Построение графика для каждой категории
for category in categories:
    plot_metrics_by_category(df, category)
```
