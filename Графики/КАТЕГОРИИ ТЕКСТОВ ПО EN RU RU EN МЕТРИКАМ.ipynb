```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#     data.csv
df = pd.read_csv('data.csv')

#    
metrics = ['BLEU', 'ROUGE-1', 'METEOR']

#   
categories = df[''].unique()

#      
def plot_metrics_by_category(df, category):
    subset = df[df[''] == category]

    directions = subset[' '].unique()
    mean_values = {}

    for direction in directions:
        mean_values[direction] = subset[subset[' '] == direction][metrics].mean()
    
    fig, ax = plt.subplots(figsize=(12, 8))

    #       
    bar_width = 0.35
    index = np.arange(len(metrics))
    colors = ['#1f77b4', '#2ca02c']  #   

    for i, direction in enumerate(directions):
        values = mean_values[direction]
        bars = ax.bar(index + i * bar_width, values, width=bar_width, label=direction, color=colors[i], edgecolor='black')

        #       
        for bar, value in zip(bars, values):
            ax.text(bar.get_x() + bar.get_width() / 2, value, f'{value:.2f}', ha='center', va='bottom', fontsize=10, color='white',
                    bbox=dict(facecolor='black', alpha=0.7, edgecolor='none', boxstyle='round,pad=0.3'))

    ax.set_xlabel('', fontsize=14)
    ax.set_ylabel(' ', fontsize=14)
    ax.set_title(f'{category} -    ', fontsize=16)
    ax.set_xticks(index + bar_width / 2)
    ax.set_xticklabels(metrics, fontsize=12)
    ax.legend(title='', fontsize=12, title_fontsize='13')

    #      
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

    plt.grid(axis='y', linestyle='--', linewidth=0.7)
    plt.tight_layout()
    plt.show()

    #    
    print(f"   '{category}':")
    for direction, values in mean_values.items():
        sorted_values = values.sort_values()
        print(f"{direction}:")
        print(sorted_values)
    print("\n")

#     
for category in categories:
    plot_metrics_by_category(df, category)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#     data.csv
df = pd.read_csv('data.csv')

#    
metrics = ['BLEU', 'ROUGE-1', 'METEOR']

#   
categories = df[''].unique()

#      
def plot_metrics_by_category(df, category):
    subset = df[df[''] == category]

    directions = subset[' '].unique()
    mean_values = {}

    for direction in directions:
        mean_values[direction] = subset[subset[' '] == direction][metrics].mean()
    
    fig, ax = plt.subplots(figsize=(10, 6))

    #       
    bar_width = 0.35
    index = np.arange(len(metrics))
    colors = ['blue', 'orange']
    
    for i, direction in enumerate(directions):
        values = mean_values[direction]
        ax.bar(index + i * bar_width, values, width=bar_width, label=direction, color=colors[i])

        #     
        for j, value in enumerate(values):
            ax.text(index[j] + i * bar_width, value, f'{value:.2f}', ha='center', va='bottom')

    ax.set_xlabel('')
    ax.set_ylabel(' ')
    ax.set_title(f'{category} -    ')
    ax.set_xticks(index + bar_width / 2)
    ax.set_xticklabels(metrics)
    ax.legend(title='')

    plt.tight_layout()
    plt.show()

    #    
    print(f"   '{category}':")
    for direction, values in mean_values.items():
        sorted_values = values.sort_values()
        print(f"{direction}:")
        print(sorted_values)
    print("\n")

#     
for category in categories:
    plot_metrics_by_category(df, category)
```
