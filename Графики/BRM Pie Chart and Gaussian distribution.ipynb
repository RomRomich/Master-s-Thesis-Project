```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Загрузка данных
df = pd.read_csv('data.csv')

# Функция для группировки и построения графиков
def plot_combined_by_category(direction):
    # Фильтрация по направлению перевода
    filtered_df = df[df['Направление перевода'] == direction]
    # Группировка по категориям и вычисление среднего значения всех метрик
    category_means = filtered_df.groupby('Категория')[['BLEU', 'ROUGE-1', 'METEOR']].mean().mean(axis=1)
    # Печать значений от меньшего к большему
    print(f"\nСредние значения всех метрик по категориям ({direction}) от меньшего к большему:")
    sorted_category_means = category_means.sort_values()
    for category, value in sorted_category_means.items():
        print(f"{category}: {value:.4f}")
    # Построение графика Pie Chart
    plt.figure(figsize=(10, 8))
    sorted_category_means.plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'Средние значения всех метрик по категориям ({direction})')
    plt.ylabel('')
    plt.show()
    # Построение графика с нормальным распределением (кривой плотности KDE)
    plt.figure(figsize=(10, 6))
    sns.kdeplot(sorted_category_means, shade=True, bw_adjust=1.5)
    plt.title(f'Нормальное распределение всех метрик по категориям ({direction})')
    plt.xlabel('Среднее всех метрик')
    plt.ylabel('Density')
    plt.show()

# Построение графиков для EN-RU и RU-EN
plot_combined_by_category('EN-RU')
plot_combined_by_category('RU-EN')
```
